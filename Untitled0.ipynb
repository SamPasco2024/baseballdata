```python
# https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LogisticRegression.html
# imports
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
# load the data from the web link 
dt = pd.read_csv("https://raw.githubusercontent.com/SamPasco2024/stockdata/main/5%20MINUTECSV.csv")

data = dt.drop('Date', axis=1)
#dataNoHoL = data.drop('NextHigherorLower', axis=1)

data
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

#train/test split
Xtrain, Xtest, Ytrain, Ytest = train_test_split(data.drop("NextHigherorLower", axis=1), data["NextHigherorLower"])
Xtrain

#FitModel
LogReg = LogisticRegression(solver="lbfgs")
LogReg.fit(Xtrain, Ytrain)

#Playground
rowNumber = 1520
samplerow = data[rowNumber:rowNumber+1].drop("NextHigherorLower", axis=1)
samplerowAsArray = np.array(samplerow)

samplerow

######
samplerowActual = data[rowNumber:rowNumber+1]["NextHigherorLower"]
print("actual:", int(samplerowActual))
ans = LogReg.predict(samplerowAsArray)[0]

print("predicted:", ans)

#Score
LogReg.score(Xtest, Ytest)
```

# 50 Moving Average

```python
newData = pd.read_csv("https://raw.githubusercontent.com/SamPasco2024/stockdata/main/5minEditedFor50MovingAvg.csv")
#train/test split
Xtrain3, Xtest3, Ytrain3, Ytest3 = train_test_split(newData.drop("NextHigherorLower", axis=1), newData["NextHigherorLower"])

#FitModel train/test
LogReg3 = LogisticRegression(solver="lbfgs")
LogReg3.fit(Xtrain3, Ytrain3)

#FitModel train/test
LogReg4 = LogisticRegression(solver="lbfgs")
LogReg4.fit(newData.drop("NextHigherorLower", axis=1), newData["NextHigherorLower"])

####

print("With train/test split:")
print(LogReg3.score(Xtest3, Ytest3))

print("Without train/test split:")
print(LogReg4.score(newData.drop("NextHigherorLower", axis=1), newData["NextHigherorLower"]))
```

# A-50Moving1Prev

```python
a50Moving = pd.read_csv("https://raw.githubusercontent.com/SamPasco2024/stockdata/main/A-50Moving1Prev.csv")
#train/test split
Xtrain50, Xtest50, Ytrain50, Ytest50 = train_test_split(a50Moving.drop("NextHigherorLower", axis=1), a50Moving["NextHigherorLower"])

#FitModel train/test
LogReg50 = LogisticRegression(solver="lbfgs")
LogReg50.fit(Xtrain50, Ytrain50)
print("with train/test")
print(LogReg50.score(Xtest50, Ytest50))


#FitModel
LogReg50w = LogisticRegression(solver="lbfgs")
LogReg50w.fit(a50Moving.drop("NextHigherorLower", axis=1), a50Moving["NextHigherorLower"])
print("without train/test")
LogReg50w.score(a50Moving.drop("NextHigherorLower", axis=1), a50Moving["NextHigherorLower"])
```
